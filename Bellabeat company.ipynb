```python
plt.figure(figsize=(15,8),dpi=80)
sns.heatmap(data=data.drop(['Id','date'],axis=1).corr(),annot=True,cmap='Blues')
plt.show()
```

Therefore from the above heatmap graph it is visible that total steps,total distance and tracker distance arer psoitively correlated whereas total sleep records and total minutes as sleep is negatively correlated.

### Total Steps vs Calories

```python
plt.scatter(data.TotalSteps,data.Calories)
plt.xlabel("TotalSteps")
plt.ylabel("Calories")
plt.title("Total steps vs Calories")
plt.show()
```

#### The above graph concludes that total steps is positively correlated to the calories burnt .

### Total sleep records vs total minutes as sleep

```python
plt.scatter(data.TotalMinutesAsleep,data.TotalSleepRecords)
plt.title("Total sleep records vs total minutes as sleep")
plt.xlabel("TotalMinutesAsleep")
plt.ylabel("TotalSleepRecords")
plt.show()
```

### Sedementary minutes vs total minute as sleep

```python
plt.scatter(data.SedentaryMinutes,data.TotalMinutesAsleep)
plt.xlabel("SedentaryMinutes")
plt.ylabel("TotalMinutesAsleep")
plt.title("Sedementary minutes vs total minute as sleep")
plt.show()
```

#### The above graph concludes that total minutes as sleep is negatively related to sedemenatry minutes .

```python
data['Day_name']=data['date'].dt.day_name()
```

```python
data
```

```python
plt.figure(figsize=(10,6),dpi=100)
data.groupby('Day_name').mean()['TotalDistance'].plot(kind='bar')
plt.xlabel("Day_name")
plt.ylabel("Total Distance")
plt.title("Distance travelled on week day")
plt.show()
```

##### Saturday is the most active day of the week and sunday is the least active day of the week

### Sleeping pattern of the week

```python
plt.figure(figsize=(10,6),dpi=100)
data.groupby('Day_name').mean()['TotalMinutesAsleep'].plot(kind='bar',color="red")
plt.xlabel("Day_name")
plt.ylabel("Total minutes asleep")
plt.title("sleeping pattern throughout the week")
plt.show()
```

* Sunday is most sleeping hours of the week whereas thursday and friday are having equivalent hours of week

```python
plt.figure(figsize=(14,10),dpi=100)
data.groupby('Day_name').mean()[['VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes']].plot(kind='area')
plt.xlabel("Day_name")
plt.ylabel("Minutes")
plt.title("Distribution of activity") 
plt.show()
```

##### How many sleeping phase or cycle are completed

```python
plt.figure(figsize=(10,6),dpi=100)
data["TotalSleepRecords"].value_counts().plot(kind="bar")
plt.xlabel("value counts")
plt.ylabel("TotalSleepRecords")
plt.title("Sleeping phase")
plt.show()
```

* From the above chart it is clear that many people are not completing their adequate amount of sleep cycle .
* The result of which is that their functioning of mind will become slow .
* Also, they will engage less in physical activity.

### Analysis of hourly_step dataset

```python
hourly_step['Time_only']=pd.to_datetime(hourly_step['ActivityHour']).dt.time
```

```python
hourly_step.head()
```

```python
hourly_step['Time_only']=hourly_step['Time_only'].apply(lambda x:str(x))
hourly_step.head()
```

```python
h_steps=pd.DataFrame(hourly_step.groupby("Time_only").mean()['StepTotal']).reset_index()
```

```python
h_steps.head()
```

##### What time of the day are people active ?

```python
plt.figure(figsize=(14,10),dpi=100)
sns.barplot(data=h_steps,x='Time_only',y='StepTotal',palette="flare")
plt.xlabel("Time_only")
plt.ylabel("StepTotal")
plt.title("Most active time")
plt.xticks(rotation=90)
plt.show()
```

* The prefered time for being active is in between 6AM to 10AM 
* And most active time is in between 12PM to 7PM
